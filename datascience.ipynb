```python
import matplotlib.pyplot as plt

tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period="1y")

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Kapanış Fiyatı')
    plt.title(title)
    plt.xlabel('Tarih')
    plt.ylabel('Fiyat (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

make_graph(tesla_data, 'Tesla Hisse Senedi Fiyatı')
```

```python
gamestop = yf.Ticker('GME')
gamestop_data = gamestop.history(period="1y")

make_graph(gamestop_data, 'GameStop Hisse Senedi Fiyatı')
```
